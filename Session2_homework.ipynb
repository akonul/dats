### 5.Illustrate relationship between Total Bags and Small Bags using scatter plot.

```python
import matplotlib.pyplot as plt
x= dat['Total Bags']
y = dat['Small Bags']
plt.scatter(x,y, label = '1st sctrplot', color ='red')
plt.legend()
```

### 7.Which type of Avocados are expensive?(with boxplot)

```python
dat.columns
```

```python
import seaborn as sns 
#plt.boxplot(dat['AveragePrice'])
sns.boxplot(data=dat, x='type', y = 'AveragePrice')
```

### 8.Construct heatmap correlation matrix.

```python
dat.columns
```

```python
fig = plt.figure(figsize=(5,5))
sns.heatmap(pd.pivot_table(data = dat, columns='type', index='year', values='AveragePrice'), annot=True, cmap='rainbow', fmt='.1f', linewidths=0.5)
```

### 9.Find relationship between large and small bags using scatterplot.(use jointplot function)

```python
sns.jointplot(data=dat, x='Large Bags', y='Small Bags', kind='kde' )
```

### 10.Find linear relationship between Total Volume and Bags for each avocado type.(use lmplot function)

```python
sns.lmplot(data=dat, x= 'Total Volume', y = 'Large Bags')
```
