**Colour abbreviation**      **Colour name**

m                               magenta

There are several ways to specify colours, other than by colour abbreviations:
    
•	The full colour name, such as yellow

•	Hexadecimal string such as ##FF00FF

•	RGB tuples, for example (1, 0, 1)

•	Grayscale intensity, in string format such as ‘0.7’.


```python
x16 = np.arange(1, 5)
# Plot green line with x16
plt.plot(x16, c='g')
# Plot red line with x16+1
plt.plot(x16+1, c='r')
# Plot blue line with x16+2
plt.plot(x16+2, c='blue')
```

## Control line styles

All the available line styles are available in the following table:


**Style abbreviation**  **Style**

-                    solid line
   
--                   dashed line
   
-.                   dash-dot line
   
:                    dotted line
   


Now, we can see the default format string for a single line plot is 'b-'.

```python
x16 = np.arange(1, 5)
# plot blue line with -- , red line with -. , green line with :
plt.plot(x16, c='g', ls = "--")
plt.plot(x16+1, c='r', ls = "-")
plt.plot(x16+2, c='b', ls = ":")
```

```python
x = np.linspace(0, 2)
# plot x and x and label it as 'linear'
plt.plot(x,x, label = 'linear')
# plot x and x**2 and label it as 'quadratic'
plt.plot(x,x**2, label = 'quadratic')
# plot x and x**3 and label it as cubic
plt.plot(x,x**3, label= 'cubic')
# label x and y axes
plt.xlabel('Xaxis')
plt.ylabel('Yaxis')
# show legend
plt.legend()
```

### 3. Find correlation between variables. Then use heatmap for visualization (annot=True - numbers will visually appear in heatmap)

```python
dat.corr()
```

```python
f,ax=plt.subplots(figsize = (5,5))
sbn.heatmap(dat.corr(),annot= True,linewidths=0.5,fmt = ".1f")
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.title('Correlation Map')
plt.savefig('graph.png')
plt.show()
```

### 4. Display relationship between highway miles per gallon and city miles per gallon

### 5. Show distributions and scatters between all variables by type of drive train

### 6. Which class has most observed in dataset. 
* Use countplot

```python
sbn.countplot(dat['manufacturer'])
sbn.countplot(dat['model'])
sbn.countplot(dat['displ'])
sbn.countplot(dat['year'])
sbn.countplot(dat['cyl'])
sbn.countplot(dat['trans'])
sbn.countplot(dat['drv'])
sbn.countplot(dat['cty'])
sbn.countplot(dat['hwy'])
sbn.countplot(dat['fl'])
sbn.countplot(dat['class'])
# midsize is the most observed one 
```

### 7. Visualize engine displacement by each class
